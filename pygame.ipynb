```python
# смысл игры: пробежать с мячом и не столкнуться с защитником
# при победе/поражении происходит автоматический перезапуск игры через 2 секунды
class Player():
    def __init__(self, screen):
        self.screen = screen
        self.image = pygame.image.load('player.png')
        self.image = pygame.transform.scale(self.image, (50, 50)) 
        self.rect = self.image.get_rect()
        self.rect.center = (250, 650)
        self.mright = False
        self.mleft = False
        self.mup = False
        self.mdown = False
    def output(self):
        self.screen.blit(self.image, self.rect)
    def update(self):
        if self.mright and self.rect.x < 450:
            self.rect.x += 5
        if self.mleft and self.rect.x > 0:
            self.rect.x += -5
        if self.mup and self.rect.y > 0:
            self.rect.y += -5
        if self.mdown and self.rect.y < 750:
            self.rect.y += +5

# класс защитник
class Defender():
    def __init__(self, x, y, screen):
        self.screen = screen
        self.image = pygame.image.load('defender.png')
        self.image = pygame.transform.scale(self.image, (60, 60)) 
        self.rect = self.image.get_rect()
        self.rect.center = (x, y)
        self.x = x
        self.step = x/15
    def output(self):
        self.screen.blit(self.image, self.rect)
    def update(self):
        self.rect.x += self.step
        if self.rect.x <= 50:
            self.step = self.x/15
        elif self.rect.x >= 450:
            self.step = -self.x/15
            
# функция игры
def my_game():
    pygame.init()
    screen = pygame.display.set_mode((500, 700))
    pygame.display.set_caption("Игра 1")
    clock = pygame.time.Clock()

    player = Player(screen)
    defender_4 = Defender(170, 100, screen)
    defender_3 = Defender(150, 250, screen)
    defender_2 = Defender(80, 400, screen)
    defender_1 = Defender(100, 550, screen)
    
    running = True
    while running:
        clock.tick(30)
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_RIGHT:
                    player.mright = True
                elif event.key == pygame.K_LEFT:
                    player.mleft = True
                elif event.key == pygame.K_UP:
                    player.mup = True
                elif event.key == pygame.K_DOWN:
                    player.mdown = True

            elif event.type == pygame.KEYUP:
                if event.key ==pygame.K_RIGHT:
                    player.mright = False
                elif event.key == pygame.K_LEFT:
                    player.mleft = False
                elif event.key == pygame.K_UP:
                    player.mup = False
                elif event.key == pygame.K_DOWN:
                    player.mdown = False

        player_center = player.rect
        
        #столкновение
        if player_center.collidepoint(defender_1.rect.center) or player_center.collidepoint(defender_2.rect.center) or player_center.collidepoint(defender_3.rect.center) or player_center.collidepoint(defender_4.rect.center):
            image_lose = pygame.image.load('you_lose.png')
            image_lose = pygame.transform.scale(image_lose, (300, 300)) 
            screen.blit(image_lose,(120,100))
            pygame.display.update()
            pygame.time.delay(2000)
            my_game()
            break
        
        #выигрыш
        if player.rect.y == 0 :
            image_lose = pygame.image.load('you_win.png')
            image_lose = pygame.transform.scale(image_lose, (300, 300)) 
            screen.blit(image_lose,(120,100))
            pygame.display.update()
            pygame.time.delay(2000)
            my_game()
            break
        player.update()
        defender_1.update()
        defender_2.update()
        defender_3.update()
        defender_4.update()
        
        image_screen = pygame.image.load('field.jpg')
        image_screen = pygame.transform.scale(image_screen, (800, 800)) 
        screen.blit(image_screen,(0, 0))
        player.output()
        defender_1.output()
        defender_2.output()
        defender_3.output()
        defender_4.output()
        pygame.display.flip()

    pygame.quit()

my_game()
```
